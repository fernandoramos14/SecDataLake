```python
import json
import random
from datetime import datetime, timedelta

# Configuración inicial
usuarios = [f"empleado.{i}@novus.com" for i in range(1, 51)]  # 50 usuarios simulados
urls_maliciosas = [
    "https://login-microsoft-verify.com/auth",
    "https://actualizacion-banco-seguro.net/login",
    "https://soporte-it-cambiodepassword.org/update"
]
urls_legitimas = ["https://linkedin.com", "https://github.com", "https://aws.amazon.com"]

email_logs = []
proxy_logs = []

# Base de tiempo para simular eventos (Mayo 2026)
base_time = datetime(2026, 5, 18, 9, 0, 0)

# 1. Generar Logs de Correo (Capa Bronze - Fuente A)
for i in range(200):  # 200 correos recibidos con códigos QR
    mail_id = f"MSG{100000 + i}"
    usuario = random.choice(usuarios)
    is_malicious = random.choices([True, False], weights=[30, 70])[0] # 30% probabilidad de ataque Qishing

    url_qr = random.choice(urls_maliciosas) if is_malicious else random.choice(urls_legitimas)
    # Variación de tiempo de llegada
    event_time = base_time + timedelta(minutes=random.randint(1, 2880)) # repartido en 2 días

    email_entry = {
        "event_timestamp": event_time.strftime("%Y-%m-%dT%H:%M:%SZ"),
        "mail_id": mail_id,
        "sender_address": "attacker@phishingnet.com" if is_malicious else "newsletter@info.com",
        "recipient_address": usuario,
        "subject": "URGENTE: Validar credenciales corporativas" if is_malicious else "Reporte Semanal Disponible",
        "qr_extracted_url": url_qr,
        "attachment_name": "qr_code_access.png" if is_malicious else "invoice.pdf"
    }
    email_logs.append(email_entry)

    # 2. Simular Comportamiento Humano/Forense (Capa Bronze - Fuente B)
    # Si era malicioso, hay un 40% de probabilidad de que el usuario haya escaneado el QR con su móvil corporativo
    if is_malicious and random.random() < 0.40:
        # El clic ocurre entre 2 y 15 minutos después de recibir el correo
        click_time = event_time + timedelta(minutes=random.randint(2, 15))

        proxy_entry = {
            "click_timestamp": click_time.strftime("%Y-%m-%dT%H:%M:%SZ"),
            "user_principal_name": usuario,
            "source_ip": f"192.168.1.{random.randint(10, 250)}",
            "destination_url": url_qr,
            "http_status": 200,
            "bytes_transferred": random.randint(2000, 8000)
        }
        proxy_logs.append(proxy_entry)

# Guardar los archivos JSON localmente
with open("email_qr_logs.json", "w") as f:
    json.dump(email_logs, f, indent=2)

with open("proxy_web_logs.json", "w") as f:
    json.dump(proxy_logs, f, indent=2)

print(f"¡Éxito! Generados {len(email_logs)} logs de correo y {len(proxy_logs)} incidentes de navegación correlacionados.")
```
